# Step - 3 - Univariate Analysis -> PDF, Histograms, Boxplots, Countplots, etc..
- Find the outliers in each numerical column



```python
num_rows = 9
num_cols = 3

fig, axes = plt.subplots(num_rows, num_cols, figsize=(30, 60))

axes_flat = axes.flatten()

for i, column in enumerate(data.select_dtypes(include=[np.number]).columns):
    sns.boxplot(y=data[column], ax=axes_flat[i])
    axes_flat[i].set_title(column)

for j in range(len(data.select_dtypes(include=[np.number]).columns), num_rows * num_cols):
    axes_flat[j].axis('off')

plt.tight_layout()
plt.show()
```

#
- Understand the probability and frequency distribution of each numerical column


```python
for column in data.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[column], bins=20, kde=True, color='skyblue', edgecolor='black',stat="probability")
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

#
- Understand the frequency distribution of each categorical Variable/Column

```python
data['DOL'].replace('present', datetime.now().strftime('%Y-%m-%d 00:00:00'),inplace=True)
for column in data.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(60, 30))
    sns.countplot(data=data, x=column, order=data[column].value_counts().index, palette='viridis')
    plt.title(f'Frequency Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right',fontsize=10)
    plt.show()
```

# Step - 4 - Bivariate Analysis
- Discover the relationships between numerical columns using Scatter plots, hexbin plots, pair plots, etc..


```python
plt.figure(figsize=(20, 15))
sns.heatmap(data.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
numerical_columns = data.select_dtypes(include=[np.number]).columns
```

```python
import itertools
combinations = list(itertools.combinations(numerical_columns, 2))
num_plots = len(combinations)
num_cols = 3  # Number of columns for subplots
num_rows = num_plots // num_cols + (1 if num_plots % num_cols > 0 else 0)  # Number of rows for subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(5*num_cols, 5*num_rows))

if num_rows == 1:
    axs = axs.reshape(1, -1)

# Create scatter plots for each combination
for i, (var1, var2) in enumerate(combinations):
    row = i // num_cols
    col = i % num_cols
    sns.scatterplot(x=var1, y=var2, data=data, ax=axs[row][col])
    axs[row][col].set_title(f'Scatter Plot: {var1} vs {var2}')

for i in range(num_plots, num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    axs[row][col].axis('off')

plt.tight_layout()

# Show the plots
plt.show()
```

```python

# Set up the figure and axes
num_plots = len(combinations)
num_cols = 3  # Number of columns for subplots
num_rows = num_plots // num_cols + (1 if num_plots % num_cols > 0 else 0)  # Number of rows for subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(5*num_cols, 5*num_rows))

# Flatten the axes if there's only one row
if num_rows == 1:
    axs = axs.reshape(1, -1)

# Create hexbin plots for each combination
for i, (var1, var2) in enumerate(combinations):
    row = i // num_cols
    col = i % num_cols
    axs[row][col].hexbin(x=data[var1], y=data[var2], gridsize=20, cmap='Blues')
    axs[row][col].set_title(f'Hexbin Plot: {var1} vs {var2}')
    axs[row][col].set_xlabel(var1)
    axs[row][col].set_ylabel(var2)
    fig.colorbar(axs[row][col].collections[0], ax=axs[row][col])

# Hide any empty subplots
for i in range(num_plots, num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    axs[row][col].axis('off')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
